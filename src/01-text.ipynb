```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
```

```python
spanish_st = stopwords.words('spanish')
print(spanish_st)
```

```python
fdist.__dir__()
```

```python
words = [(word.lower(), count) for word, count in fdist.items() if word not in spanish_st]
words
```

```python
dtypes = [('word', 'S10'), ('frequency', int)]
palabras_interesantes = np.array(words, dtype=dtypes)
```

```python
palabras_interesantes = np.sort(palabras_interesantes, order='frequency')
palabras_interesantes
```

```python
top_words = 20
section = palabras_interesantes[-top_words:]
x = np.arange(len(section))
y = [freq[1] for freq in section]
plt.figure(figsize=(10,5))
plt.plot(x, y)
plt.xticks(x, [word[0] for word in section], rotation=45)
plt.grid(True)
plt.show()
```
